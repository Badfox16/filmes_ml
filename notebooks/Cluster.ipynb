```python
from sklearn.datasets import load_digits
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
# Carregar o conjunto de dados de dígitos
digits = load_digits()
X = digits.data  # Características dos dígitos
y = digits.target  # Rótulos verdadeiros dos dígitos
```

```python
# Reduzir a dimensionalidade para facilitar a visualização
pca = PCA(n_components=2)
X_reduced = pca.fit_transform(X)
```

```python
# Instanciar o modelo K-means com 10 clusters (número de dígitos)
kmeans = KMeans(n_clusters=10, random_state=42)

# Ajustar o modelo aos dados
kmeans.fit(X)

# Obter os rótulos de cluster para cada ponto de dado
labels = kmeans.labels_
```

```python
# Plotar os clusters
plt.figure(figsize=(8, 6))

for i in range(10):
    plt.scatter(X_reduced[labels == i, 0], X_reduced[labels == i, 1], label=f'Cluster {i}', alpha=0.5)

plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.title('Clustering dos dígitos escritos à mão')

plt.legend()
plt.show()
```

**IRIS**

```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Carregar o conjunto de dados Iris
iris = load_iris()
X = iris.data  # Características das flores

# Vamos utilizar apenas duas características para facilitar a visualização
X = X[:, :2]

# Instanciar o modelo K-means com 3 clusters (sabemos que há 3 espécies de flores)
kmeans = KMeans(n_clusters=3)

# Ajustar o modelo aos dados
kmeans.fit(X)

# Obter os rótulos de cluster para cada ponto de dado
labels = kmeans.labels_

# Obter os centróides de cada cluster
centroids = kmeans.cluster_centers_

# Plotar os clusters e os centróides
plt.figure(figsize=(8, 6))

plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='x', s=200, label='Centroids')

plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1])
plt.title('Clustering das flores Iris')

plt.legend()
plt.show()
```
