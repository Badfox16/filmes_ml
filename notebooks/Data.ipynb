```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# Gerar dados de exemplo
np.random.seed(0)
X = np.random.rand(100, 1) * 10 
y = 2 + 3 * X + np.random.randn(100, 1) * 2  
```

```python

# Dividir os dados em conjuntos de treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Criar e treinar o modelo de regressão linear
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Fazer previsões no conjunto de teste
y_pred = model.predict(X_test)

# Calcular o erro quadrático médio (MSE) das previsões
mse = mean_squared_error(y_test, y_pred)
print('Erro Quadrático Médio (MSE):', mse)
```

```python
# Visualizar os resultados
plt.scatter(X_test, y_test, color='black')
plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.xlabel('Tamanho do Imóvel')
plt.ylabel('Preço do Imóvel')
plt.title('Regressão Linear - Previsão de Preços Imobiliários')
plt.show()
```

Com arrays basicos

```python
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
x = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
y = np.array([2, 4, 5, 4, 5])
```

```python
reg = LinearRegression()
reg.fit(x,y)
```

```python
x_novo= np.array([[6]])
y_novo = reg.predict(x_novo)
```

```python
print(f"Previsão para x=6: y = {y_novo[0]:.2f}")
```

```python
plt.scatter(x,y, color='red')
plt.scatter(x_novo, y_novo, color='blue')
plt.xlabel('Eixo de X')
plt.ylabel('Eixo de Y')
plt.show
```
